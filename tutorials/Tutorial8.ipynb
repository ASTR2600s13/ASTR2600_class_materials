Finally, some cool trickery with internet access

````python
# run this cell
import urllib    # this is used for internet access
import StringIO  # this is used to turn a "string" into readable binary data
import PIL       # PIL is the "Python Imaging Library" for working with images
# grab a cool image from APOD
f = urllib.urlopen("http://apod.nasa.gov/apod/image/1301/OrionGEMS_FINAL.jpg").read()
# Open it
img = PIL.Image.open(StringIO.StringIO(f))
# and display it
imshow(img)                    
````

You can turn the image into an array:

````python
imgarr = np.array(img)
# array.shape shows the dimensions of the array
imgarr.shape
````

There are 3 color "planes" in the image, RGB.  You can show each of them individually:

````python
imshow(imgarr[:,:,0])
````

Do the same thing with the other layers:

Now, show the histograms of each of these layers.  I'll give you one example, because if you tried to do `hist(imgarr[:,:,0])`, you'd end up with 2000 overplotted histograms!  Instead, you need to "flatten" the data so it gets treated as one array:

````python
# the "alpha" value controls the "transparency" of the color
hist(imgarr[:,:,0].flat, alpha=0.5, color='r')
````
